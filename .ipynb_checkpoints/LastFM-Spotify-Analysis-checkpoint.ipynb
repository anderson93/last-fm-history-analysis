```python
import time
import sys
import datetime as dt
import numpy as np

fail_count = 0
sectry = 0
results = []

for i, (artist, album, track) in enumerate(music_data.values):
    start = dt.datetime.utcnow()

    try:
        track_uri = sp.search(q=f'artist:{artist} track:{track} album:{album}', type='track', market='BR', limit=1)
        time.sleep(np.random.random(3).sum())
    except:
        try:
            track_uri = sp.search(q=f'artist:{artist} track:{track}', type='track', market='BR', limit=1)
            time.sleep(np.random.random(3).sum())
            sectry += 1
        except:
            track_uri = 0
            fail_count += 1

    results.append(track_uri)
    timer = dt.datetime.utcnow() - start
    toend = timer*(len(music_data)-i)
    stats = 100.0*i/len(music_data)
    if i==0:
        elapsed = timer
    else:
        elapsed += timer
    sys.stdout.write(f'\r {stats} %, 2ºtry: {sectry}, nans:{fail_count}, this: {timer}, elapsed: {elapsed}, to end: {toend}')
    print(f'\r {stats} %, 2ºtry: {sectry}, nans:{fail_count}, this: {timer}, elapsed: {elapsed}, to end: {toend}', file=open('stats.txt', 'w'))
```

```python
def just_the_items(result):
    try:
        return result['tracks']['items'][0]
    except :
        pass

def searches_not_empty(result):
    try:
        result['album']
        return result
    except:
        pass
    
items_list = list(map(lambda x: searches_not_empty(just_the_items(x)), results))

music_data_spotify = {}
music_data_spotify['album_name'] = []
music_data_spotify['album_release_date'] = []
music_data_spotify['album_total_tracks'] = []
music_data_spotify['artists'] = []
music_data_spotify['artists_number'] = []
music_data_spotify['disc_number'] = []
music_data_spotify['duration_ms'] = []
music_data_spotify['explicit'] = []
music_data_spotify['name'] = []
music_data_spotify['popularity'] = []
music_data_spotify['track_number'] = []
music_data_spotify['uri'] = []

for x in items_list:
    try:
        music_data_spotify['album_name'].append(x['album']['name'])
    except:
        music_data_spotify['album_name'].append('unknown')
    try:
        music_data_spotify['album_release_date'].append(x['album']['release_date'])
    except:
        music_data_spotify['album_release_date'].append('unknown')
    try:
        music_data_spotify['album_total_tracks'].append(x['album']['total_tracks'])
    except:
        music_data_spotify['album_total_tracks'].append('unknown')
    try:
        music_data_spotify['artists'].append(x['artists'])
    except:
        music_data_spotify['artists'].append('unknown')
    try:
        music_data_spotify['artists_number'].append(len(x['artists']))
    except:
        music_data_spotify['artists_number'].append('unknown')
    try:
        music_data_spotify['disc_number'].append(x['disc_number'])
    except:
        music_data_spotify['disc_number'].append('unknown')
    try:
        music_data_spotify['duration_ms'].append(x['duration_ms'])
    except:
        music_data_spotify['duration_ms'].append('unknown')
    try:
        music_data_spotify['explicit'].append(x['explicit'])
    except:
        music_data_spotify['explicit'].append('unknown')
    try:
        music_data_spotify['name'].append(x['name'])
    except:
        music_data_spotify['name'].append('unknown')
    try:
        music_data_spotify['popularity'].append(x['popularity'])
    except:
        music_data_spotify['popularity'].append('unknown')
    try:
        music_data_spotify['track_number'].append(x['track_number'])
    except:
        music_data_spotify['track_number'].append('unknown')
    try:
        music_data_spotify['uri'].append(x['uri'])
    except:
        music_data_spotify['uri'].append('unknown')
        
```

```python
music_data_spotify_df = pd.DataFrame(music_data_spotify)
music_data_spotify_df = music_data_spotify_df[music_data_spotify_df['uri'] != 'unknown'].drop_duplicates('uri').reset_index(drop=True).copy()
```

```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]

spotify_uri = music_data_spotify_df.uri.values.tolist()

audio_list_uri = list(chunks(spotify_uri, 50))
audio_list_uri[:2]
```

```python
import time
import sys

audio_result = []
sectry = 0
fail_count = 0
for i, search in enumerate(audio_list_uri):
    start = dt.datetime.utcnow()
    
    try:
        time.sleep(np.random.random(3).sum())
        audio_result += sp.audio_features(tracks=search)
    except:
        try:
            time.sleep(np.random.random(3).sum())
            audio_result += sp.audio_features(tracks=search)
            sectry += 1
        except:
            time.sleep(np.random.random(3).sum())
            audio_result += sp.audio_features(tracks=search)
            fail_count += 1

    timer = dt.datetime.utcnow() - start
    toend = (timer + dt.timedelta(seconds=3))*(len(audio_list_uri)-i)
    stats = 100.0*i/len(audio_list_uri)
    if i==0:
        elapsed = timer
    else:
        elapsed += timer
    sys.stdout.write(f'\r {stats} %, 2ºtry: {sectry}, nans:{fail_count}, this: {timer}, elapsed: {elapsed}, to end: {toend}')
    print(f'\r {stats} %, 2ºtry: {sectry}, nans:{fail_count}, this: {timer}, elapsed: {elapsed}, to end: {toend}', file=open('stats-audio.txt', 'w'))
```

```python
from sklearn.preprocessing import RobustScaler

rs = RobustScaler(quantile_range = (0.2, 0.8))

X = rs.fit_transform(cleaned_data.drop_duplicates(subset = ['id', 'artist', 'track'])[['acousticness', 'danceability', 'duration_ms', 'energy',
                                                                                           'instrumentalness', 'key', 'liveness', 'loudness', 'mode',
                                                                                           'speechiness', 'tempo', 'time_signature', 'valence']])
```

```python
from sklearn.cluster import KMeans

clustering = {}

for x in range(2, 25):
    
    model = KMeans(n_clusters=x, n_jobs=-1, verbose=1, random_state=0)
    results = model.fit_predict(X)
    clustering[x] = [model, results, model.inertia_, model.cluster_centers_]
    print(f"{x} done!")
```

```python
inertia = [(x, clustering[x][2]) for x in clustering.keys()]
score_df = pd.DataFrame(inertia, columns=['n_clusters','inertia']).set_index('n_clusters')
score_df.plot(figsize=(12,8), xticks=range(2, 25)).grid()
```

```python
from sklearn.metrics import silhouette_score

score_df['silhouette_score'] = [silhouette_score(X, clustering[x][0].labels_) for x in clustering.keys()]
```

```python
score_df.silhouette_score.plot(figsize=(12,8), xticks=range(2, 25)).grid()
```
